```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
data={'Height': [2,3,4,5],
    'Weight': [1,2,3,4]}
df=pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)
X = df.values
```

```python
pca = PCA(n_components=1)
X_reduced = pca.fit_transform(X)
```

```python
# 4. Results
print("Reduced Data (1D Coordinates):")
print(X_reduced.flatten())

print(f"\nVariance Explained: {pca.explained_variance_ratio_[0] * 100:.2f}%")
```
