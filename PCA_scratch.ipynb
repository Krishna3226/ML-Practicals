```python
import numpy as np
import pandas as pd
```

```python
data={'Height': [2,3,4,5],
    'Weight': [1,2,3,4]}
df=pd.DataFrame(data)
```

```python
P=df.to_numpy()
```

```python
mean_p=np.mean(P,axis=0)
P_centered=P-mean_p
```

```python
cov_m=np.cov(P_centered.T)
```

```python
eigenval, eigenvect = np.linalg.eig(cov_m)
```

```python
idx = np.argsort(eigenval)[::-1]
components = eigenvect[:, idx].T
explained_var = eigenval[idx]
```

```python
P_reduced = np.dot(P_centered, components[0].T)

print("Reduced Data: ", P_reduced)
```
