```python
df=df.drop(columns=["Reconstruction Costs ('000 US$)","Insured Damage ('000 US$)","Total Damage ('000 US$)"])
```

```python
damage_cols=["Reconstruction Costs, Adjusted ('000 US$)","Insured Damage, Adjusted ('000 US$)","Total Damage, Adjusted ('000 US$)"]

for col in damage_cols:
  df[col+'_reported']=df[col].notna().astype(int)
  df[col]=df[col].fillna(0)
```

```python
df['Event Name']=df['Event Name'].fillna('Unknown')
df['Magnitude']=df['Magnitude'].fillna(0)
df['Origin']=df['Origin'].fillna('Unknown')
df['Location']=df['Location'].fillna('Unknown')

cols=['No. Injured','No. Affected','Total Deaths','Total Affected']
for col in cols:
  df[col]=df[col].fillna(0)

duration_cols = [
    'Start Month', 'Start Day',
    'End Month', 'End Day'
]

df = df.drop(columns=duration_cols)
```

```python
for col in cols:
  df[col]=df[col].astype(float)
```

```python
from scipy import stats

def calculate_statistics(data):
    data = np.array(data)

    mean = np.mean(data)
    mode_res = stats.mode(data, keepdims=True)
    mode = mode_res.mode[0]
    variance = np.var(data)
    std_dev = np.std(data)

    data_sorted = np.sort(data)

    median = np.median(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    print(data)
    print(f"Mean:               {mean:.2f}")
    print(f"Median:             {median:.2f}")
    print(f"Mode:               {mode}")
    print(f"Variance:           {variance:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Quartile 1 (Q1):    {q1:.2f}")
    print(f"Quartile 3 (Q3):    {q3:.2f}")
    print(f"Interquartile Range:{iqr:.2f}")

calculate_statistics(x)
```
